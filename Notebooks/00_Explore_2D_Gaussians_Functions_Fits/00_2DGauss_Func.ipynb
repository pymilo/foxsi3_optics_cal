## Most Simple 2D Gaussian by defining customed function
<div class="alert alert-info">
    Goal: Define a Function that returns a 2D Gaussian and plot it.
</div>

### Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
## uncomment this line if you want to rotate the plot within the notebook.
%matplotlib notebook 
```

### Workspace

```python
## Function definition:
def gaussian(height, center_x, center_y, width_x, width_y):
    """Returns a gaussian function with the given parameters"""
    width_x = float(width_x)
    width_y = float(width_y)
    return lambda x,y: height*np.exp(
                -(((center_x-x)/width_x)**2+((center_y-y)/width_y)**2)/2)

# Create the gaussian data
Xin, Yin = np.mgrid[0:201, 0:201]
gdata = gaussian(30, 100, 100, 20, 20)(Xin, Yin) + np.random.random(Xin.shape)

## Plot with matplotlib:
fig = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
ax.plot_surface(Xin,Yin,gdata,cmap='viridis')
plt.show()
```
